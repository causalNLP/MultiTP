```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Preference in favor of sparing characters

```python
pref=pd.read_csv("./human_char_pref_scores.csv")
```

```python
pref=pref.sort_values('gpt4_estimates',ascending=False)
```

```python
ax=sns.barplot(y=pref.CharacterType,x=pref.gpt4_estimates,orient='h',palette=['#037bfc' if v >= 0 else '#8e10c4' for v in pref.gpt4_estimates])
#sns.set_style("whitegrid")
plt.xlabel("Preference Score")
plt.ylabel("Characters")
plt.title("Preference in Favor of Sparing Characters")
ax.grid(axis='y',visible=True)
ax.grid(axis='x',visible=False)
sns.despine(top=True, right=True,left=True, bottom=True )

plt.savefig('PREFERENCES.png', transparent=True,bbox_inches='tight',dpi=250)
```
